__Machine learning__ is the field that gives to computers the ability to learn without being explicity programed (Arthur Samuel 1959). More modern definition is *"A computer program is said to learn form experience E with respect to some class of tasks T and performance measure P, if its pereformance at tasks in T, as measured by P improves with experience E"(Tom Mitchell 1998)*.

Machine learning (ML) is a large and diverse field, there is several types of Machine learning. Today's most important area is Deep Learning, which is the application to learning tasks of Artificial Neuro Networks.

There is several types of ML Algorithms: 
* __Supervised learning__ : The computer is given a certain number of pairs  input/output, by a "teacher", with the objective to find a general rule that allows to predict the output of a given input
* __Unsupervised learning__ : No labels are given to the learning algorithm, leaving it on its own to find structure in its input. Unsupervised learning can be a goal in itself (discovering hidden patterns in data) or a means towards an end
* __Reinforcement learning__ : The program runs in a dynamic enviroment in which it has to perform a certian goal (win a game, drive a car...) The program is given rewards in terms of rewards or punishements while he navigates in the problem space.

Another categorization of machine learning tasks arises when one considers the *desired output* of a machine-learned system:
* __Classification__ : a process that identifyis to which set of categories a certain in put is more likely to be part of. Exemple classification Brest cancer.
* __Regression__ : It's a statistical process for estimating the relationship between variables. Exemple y = ax+b

### Parameter learning

__ Gradient decent__ is an algorithm that minimises the Cost function J(θ0, θ1) value

How? We set random values at θ0, θ1 and we keep changing θ0, θ1 until we find the minimum

So it's simply a matter of tunning. We will see in the next videos how to compute the functions who gives us the direction to apply the correct simuntenious changes to θ0, θ1 until finding the minimum.
